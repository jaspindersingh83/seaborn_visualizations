```python
import pandas as pd
titanic=pd.read_csv("train.csv")
titanic=titanic[['Survived','Pclass','Sex','Age','SibSp','Parch','Embarked','Fare']].dropna()
titanic.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ax=sns.distplot(titanic['Age'])
plt.show()
```

```python
###Using despine and ste_style functions

ax=sns.kdeplot(titanic['Age'],shade=True,legend=False)
sns.set_style('white')
sns.despine(left=True, bottom=True)
ax.set(xlabel='Age')
plt.show()
```

```python
###Using condition/filters in a distribution
ax=sns.FacetGrid(titanic,col="Pclass",size=6)
ax.map(sns.kdeplot,'Age',shade=True)
sns.despine(bottom=True,left=True)
plt.show()
```

```python
### Using multiple condition for plotting 
ax=sns.FacetGrid(titanic,col='Survived',row='Pclass',hue='Sex',size=3)
ax.map(sns.kdeplot,'Age',shade=True,legend=True)
sns.despine(bottom=True,left=True)
plt.show()
```

```python
ax=sns.FacetGrid(titanic,col="Survived",row='Pclass',hue='Sex',size=3)
ax.map(sns.kdeplot,"Age")
sns.despine(bottom=True,left=True)
plt.show()
```
